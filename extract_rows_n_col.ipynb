```python
##to divide the total no. of frames into 15 frames, excluding the last frame

track_id_file = pd.read_csv('/home/student/Downloads/stanford_drone_dataset/annotations/bookstore/video1/Track_IDs/track_id100th.csv')
print(track_id_file.shape)

BB_x_min=track_id_file['x_min']
BB_x_max=track_id_file['x_max']
BB_y_min=track_id_file['y_min']
BB_y_max=track_id_file['y_max']

unique_frames = track_id_file['frame'].unique()
#print(unique_frames[-2])
divided_frames = np.linspace(unique_frames[0],unique_frames[-1],num=15,dtype=int,endpoint =False)
print(divided_frames)

x_center =[]
y_center =[]
width=[]
height=[]
   

for i in range(len(BB_x_min)):
    for j in divided_frames:
        if track_id_file['frame'][i]==j:
            x_center.append((BB_x_min[i] + BB_x_max[i])/2)
            y_center.append((BB_y_min[i] + BB_y_max[i])/2)
            width.append(BB_x_max[i] - BB_x_min[i])
            height.append(BB_y_max[i] - BB_y_min[i])
            
BB_detail = list(zip(x_center,y_center,width,height))
#print(BB_detail)
df1=pd.DataFrame(BB_detail,columns=['x_center','y_center','width','height'])
            

print(df1)   

    
    
    
    
    
```

```python
##working code to get 60th frame of all the videos

import pandas as pd
import numpy as np

dict = {'name': ['bookstore', 'coupa', 'deathCircle', 'gates', 'hyang', 'little', 'nexus', 'quad'],
        'number_of_videos': [7, 4, 5, 9, 15, 4, 12, 4]}
column_header = ['Track_ID','x_min','y_min','x_max','y_max','frame','lost','occluded','generated','label']

def sort_list(file):
    # to get list from dataframe column
    list_col_frame = file['frame'].tolist()
    # to arrange it in ascending order
    list_col_frame.sort()
    return list_col_frame

def select_frame(list_of_frame):
    # to get the details of every 60th frame
    l = np.arange(list_of_frame[0], list_of_frame[-1], 60)
    return l


for k in range(len(dict['name'])):
    for j in range(dict['number_of_videos'][k]):
        current_file= pd.read_csv(f'/home/shefali/Downloads/annotations/{dict["name"][k]}/video{j}/updated.csv')
        list_col_frame = sort_list(current_file)
        l = select_frame(list_col_frame)

        for i in l:
            data=current_file.loc[(current_file['frame']== i)]
            selected_frame_df=pd.DataFrame(data)
            selected_frame_df.to_csv(f'/home/shefali/Downloads/annotations/{dict["name"][k]}/video{j}/frames/frame{i}th.csv',header=column_header,index=None)
            #print(data.shape)
```
