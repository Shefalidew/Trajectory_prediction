```python
##working code to convert (to generate updated.csv file)

import csv
import numpy as np
import pandas as pd

dict = {'name': ['bookstore', 'coupa', 'deathCircle', 'gates', 'hyang', 'little', 'nexus', 'quad'],
        'number_of_videos': [7, 4, 5, 9, 15, 4, 12, 4]}

column_header = ['Track_ID','x_min','y_min','x_max','y_max','frame','lost','occluded','generated','label']

def isfloat(value):
    try:
        float(value)
        return True
    except:
        return False


for k in range(len(dict['name'])):
    for j in range(dict['number_of_videos'][k]):
        with open(f'''/home/student/Downloads/stanford_drone_dataset/annotations/{dict["name"][k]}/video{j}/annotations.csv''', 'r') as f:
            data = csv.reader(f)
            y = []
            for row in data:
                x = row[0].split()
                
                for i in range(len(x)):
                    if isfloat(x[i]) == True:
                        x[i] = int(float(x[i]))
                    elif isinstance(x[i], str) == True:
                        x[i] = x[i].replace('" "', '')
                y.append(x)
        
        my_df=pd.DataFrame(y)
        my_df.to_csv(f'/home/student/Downloads/stanford_drone_dataset/annotations/{dict["name"][k]}/video{j}/updated.csv',header=column_header,index=None)
        
```
